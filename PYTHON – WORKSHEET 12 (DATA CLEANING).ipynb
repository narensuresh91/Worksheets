```python
#Q3. Display all row index with all null values
empty_list=data[data.isnull().all(axis=1)].index
empty_list
```

```python
print(data[data.isnull().all(axis=1)].shape)
print(data.shape)
```

```python
#Q4. Determine total number of empty rows and their percentage
(131/23678)
#Removing the empty rows
indexes_to_keep = set(range(data.shape[0])) - set(empty_list)
data= data.take(list(indexes_to_keep))
```

```python
data.shape
```

```python
#Q5. Column names having missing values are:
data.isnull().sum()>0
```

```python
data.isnull().sum()
```

```python
#Q6. Missing values and percentages in each column
cols=data.columns
for i in cols:
    if data[i].isnull().sum()>0:
        print(i)
        print(data[i].isnull().sum())
        print((data[i].isnull().sum()/23547)*100,'%')
```

```python
#Q7 The row indexes having missing values
missing=data[data.isnull().any(axis=1)].index
missing
```

```python
#Q8 Missing values and percentage per row
for i in missing:
    print("Index -",i," No of missing -  ",data.iloc[i].isnull().sum(),"Percentage",(data.iloc[i].isnull().sum()/21)*100)
```

```python
#Q9. Removing the columns which have more than 30% null values:
for i in data.columns:
    if ((data[i].isnull().sum()/len(data))*100 >30):
        print(i)
        data.drop(i,axis=1,inplace=True)
```

```python
#Q10 Number and percentage of rows where 5 or more values are missing
for i in missing:
    print("Index -",i," No of missing -  ",data.iloc[i].isnull().sum()>=5)
```
